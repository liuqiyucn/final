```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib.animation import PillowWriter
import numba
from numba import jit
from scipy.linalg import eigh_tridiagonal
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

The Schrodinger's Equation is given by
$$ i\hbar \frac{\partial \psi(x,t)}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \psi(x,t)}{\partial x^2} + V(x)\psi(x,t) \hspace{20mm} \psi(-L, t) = \psi(L,t) = 0$$

To make thing easier, we set 

$$ \hbar = 1 \hspace{10mm} m = 1 \hspace{10mm} L = 1$$

With this setup, our equation becomes 

$$ i \frac{\partial \psi(x,t)}{\partial t} = -\frac{1}{2} \frac{\partial^2 \psi(x,t)}{\partial x^2} + V(x)\psi(x,t) \hspace{20mm} \psi(-1, t) = \psi(1,t) = 0$$

Applying the finite difference approach, we have the following approximation

$$ \frac{\partial \psi(x,t)}{\partial t} = \frac{\psi_i(n+1) - \psi_i(n)}{dt} \\ \frac{\partial^2 \psi(x,t)}{\partial x^2} = \frac{\psi_{i+1}(n) -2\psi_j(n) + \psi_{j-1}(n)}{d x^2}$$

Using these finite difference approximation and plugging them back to the equation, we get an explicit equation for the wavefunction at a later time

$$\psi_i(n+1) = \frac{idt}{2}\frac{\psi_{i+1}(n)+\psi_{i-1}(n)-2\psi_i(n)}{dx^2} + \psi_i(n) - idtV(x)\psi_i(n)$$

Step 1: Discretization of time, space and potential

```python
# discretize positive and time
x_pnts = 301
t_pnts = 500000
dx = 1/(x_pnts-1)
dt=1e-7
x = np.linspace(-1, 1, x_pnts)

mu, sigma = 0, 0.1
psi0 = 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-1/2 * ( (x-mu)/sigma )**2 )

mu, sigma = 0, 1/20
#V = -1e4*np.exp(-(x-mu)**2/(2*sigma**2))
V = np.zeros(x_pnts)
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,V)
plt.xlabel('$x$')
plt.ylabel('$V(x)$')
plt.show()
```

```python
plt.plot(x, psi0)
plt.show()
```

```python
psi = np.zeros([t_pnts,x_pnts]).astype(complex)
psi[0] = psi0
```

step 2: finite difference method

```python
@numba.njit
def compute_psi(psi):
    for t in range(0, t_pnts-1):
        for i in range(1, x_pnts-1):
            psi[t+1][i] = 1j/2 * dt/dx**2 * (psi[t][i+1] - 2*psi[t][i] + psi[t][i-1]) + psi[t][i] - 1j*dt*V[i]*psi[t][i]
        
        normal = np.sum(np.absolute(psi[t+1])**2)*dx
        for i in range(1, x_pnts-1):
            psi[t+1][i] = psi[t+1][i]/normal
        
    return psi
```

```python
psi_m1 = compute_psi(psi)
```

```python
plt.plot(x, np.absolute(psi_m1[250000])**2)
```
